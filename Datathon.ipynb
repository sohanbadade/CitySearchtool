```
import pandas as pd
import numpy as np
import plotly as py
import plotly.graph_objects as go
```

```
whichdf = input("Are you interested in living in just the US or globally? (Enter 'US' or 'Global')")
if (whichdf == 'Global'):
  print("Rate the following on a scale from 0-3 \n 0 = Not at all, 1 = Slightly Important, 2 = Moderately Important, 3 = Very Important" )
  valA = int(input("On a scale from 0-3: How important is cappuccino price?"))
  valB = int(input("On a scale from 0-3: How important is cinema price?"))
  valC = int(input("On a scale from 0-3: How important is wine price?"))
  valD = int(input("On a scale from 0-3: How important is gasoline?"))
  valE = int(input("On a scale from 0-3: How important is average rent?"))
  valF = int(input("On a scale from 0-3: How important is average disposable income?"))
  valG = int(input("On a scale from 0-3: How important is movehub rating?"))
  valH = int(input("On a scale from 0-3: How important is purchase power?"))
  valI = int(input("On a scale from 0-3: How important is healthcare?"))
  valJ = int(input("On a scale from 0-3: How important is pollution?"))
  valK = int(input("On a scale from 0-3: How important is quality of life?"))
  valL = int(input("On a scale from 0-3: How important is crime rating?"))
  df = pd.read_excel('/content/CitiesD (2).xlsx')

  from sklearn.preprocessing import StandardScaler
  hold = df.pop('Dummy')
  hold['City'] = df.pop('City')
  hold['lat'] = df.pop('lat')
  hold['lon'] = df.pop('lon')
  features = df.drop(columns=['Country'])

  from sklearn.preprocessing import StandardScaler
  Standard = pd.DataFrame(StandardScaler().fit_transform(features),columns = features.columns)
  
  Standard.columns = ["Cappucino","Cinema","Wine","Gasoline","Avg Rent",
                      "Avg Disposable Income","MoveHub Rating",
                      "Purchase Power","Health Care","Pollution","Quality of Life", "Crime Rating"]
  Standard['Lat'] = hold['lat']
  Standard['Long'] = hold['lon']
  Standard['City'] = hold['City']
  Standard['Score'] = (Standard['Cappucino'] * -1 * valA) + (Standard['Cinema'] * -1 * valB) + (Standard['Wine'] * -1 * valC) + (Standard['Gasoline'] * -1 * valD) + (Standard['Avg Rent'] * valE) + (Standard['Avg Disposable Income'] * valF) + (Standard['MoveHub Rating'] * valG) + (Standard['Purchase Power'] * valH) + (Standard['Health Care'] * valI) + (Standard['Pollution'] * -1 * valJ) + (Standard['Quality of Life'] * valK)+ (Standard['Crime Rating'] * -1 * valL)
  Standard['Rank'] = Standard['Score'].rank(ascending = False) 

  top10 = []

  for row, city in Standard.iterrows():
    if (city['Rank']<=10):
        top10.append(city['City'])
  
  print("Here is a list of your top 10 cities:")
  print(top10)
  print("Here is your personalized map of cities! The color field is the rank of each city based on your preferences")
  import plotly.express as px
  df = px.data.gapminder()
  fig = px.scatter_geo(df, lon = Standard['Long'], lat = Standard['Lat'], locationmode='country names', color = Standard['Rank'])
  fig.show()

else:
  print("Rate the following on a scale from 0-3 \n 0 = Not at all, 1 = Slightly Important, 2 = Moderately Important, 3 = Very Important" )
  valA = int(input("On a scale from 0-3: How important is having a large population?"))
  valB = int(input("On a scale from 0-3: How important is higher education?"))
  valC = int(input("On a scale from 0-3: How important is the average price of a single family home?"))
  valD = int(input("On a scale from 0-3: How important is the average price to rent?"))
  valE = int(input("On a scale from 0-3: How important is the percentage of mental health issues?"))
  valF = int(input("On a scale from 0-3: How important is the percentage of physical health issues?"))
  valG = int(input("On a scale from 0-3: How important is the obesity rate?"))
  valH = int(input("On a scale from 0-3: How important is walkability?"))
  valI = int(input("On a scale from 0-3: How important is transit rating?"))
  valJ = int(input("On a scale from 0-3: How important is bikability?"))
  df = pd.read_excel('/content/datathon_cleaner.xlsx')


  from sklearn.preprocessing import StandardScaler
  hold = df.pop('Dummy')
  hold['City'] = df.pop('City')
  hold['lat'] = df.pop('Lat')
  hold['lon'] = df.pop('Long')
  features = df.drop(columns=['State'])
  from sklearn.preprocessing import StandardScaler
  Standard = pd.DataFrame(StandardScaler().fit_transform(features),columns = features.columns)
  
  Standard.columns = ["Population","Higher Education","Single Family Home","Rental","Mental Health",
                      "Physical Health","Obesity",
                      "Walkability","Transit Rating","Bikability"]
  Standard['Lat'] = hold['lat']
  Standard['Long'] = hold['lon']
  Standard['City'] = hold['City']
  Standard['Score'] = (Standard['Population'] * valA) + (Standard['Higher Education'] * valB) + (Standard['Single Family Home'] * -1 * valC) + (Standard['Rental'] * -1 * valD) + (Standard['Mental Health'] * -1 * valE) + (Standard['Physical Health'] * -1 * valF) + (Standard['Obesity'] * -1 * valG) + (Standard['Walkability'] * valH) + (Standard['Transit Rating'] * valI) + (Standard['Bikability'] * valJ)
  Standard['Rank'] = Standard['Score'].rank(ascending = False) 

  top10 = []

  for row, city in Standard.iterrows():
    if (city['Rank']<=10):
        top10.append(city['City'])
  
  print("Here is a list of your top 10 cities:")
  print(top10)
  print("Here is your personalized map of cities! The color field is the rank of each city based on your preferences")
  import plotly.express as px
  df = px.data.gapminder()
  fig = px.scatter_geo(df, lon = Standard['Long'], lat = Standard['Lat'], locationmode='USA-states', color = Standard['Rank'], scope = 'usa')
  fig.show()

 
```
